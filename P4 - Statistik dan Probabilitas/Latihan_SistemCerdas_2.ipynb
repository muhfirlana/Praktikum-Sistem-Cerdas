```python
import matplotlib.pyplot as plt
```

```python
p = [2,4,4,4,5,5,7,9]
q = [10,13,15,14,13,16,18,21]
x = [1,2,3,4,5,6,7,8]
```

```python
plt.scatter(x,p,color='r',marker='d')
```

```python
plt.scatter(x,p,color='r',marker='d')
plt.scatter(x,q,color='b',marker='*')
plt.title('Correlation between p and q')
plt.xlabel('range')
plt.ylabel('value')
plt.legend((p,q),loc='lower right')
plt.show()
```
