```python
corpus = [
    'Book 1', 'Book 3','Book 4','Book 5','Book 6','Book 7','Book 8','Book 9','Book 10',
    'Book 11', 'Book 12','Book 13','Book 14','Book 15','Book 16','Book 17','Book 18','Book 19','Book 20'
]

corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer_X = vectorizer.fit_transform(corpus).todense()
vectorizer_X
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

for i in range(len(vectorizer_X)) :
    for j in range(i, len (vectorizer_X)):
        if i == j:
            continue
        jarak = euclidean_distances(vectorizer_X[i],vectorizer_X[j])
        print(f'Jarak dokumen {i+1} dan {j+1}: {jarak}')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')
vectorizer_X = vectorizer.fit_transform(corpus).todense()
vectorizer_X
```
