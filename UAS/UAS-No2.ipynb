```python
import pandas as pd
import numpy as np
```

```python
books_df = pd.read_csv('E:/!!!STIMATA/MATKUL/SISTEM CERDAS/UAS/dataset/Books.csv',usecols=['ISBN','Book-Title'],dtype={'ISBN':'str', 'Book-Title':'str'})
```

```python
rating_df = pd.read_csv("E:/!!!STIMATA/MATKUL/SISTEM CERDAS/UAS/dataset/Ratings.csv",usecols=['User-ID','ISBN','Book-Rating'],dtype={'User-ID':'str','ISBN':'str','Book-Rating':'float32'})
```

```python
books_df.head()
```

```python
books_df.shape
```

```python
rating_df.shape
```

```python
df = pd.merge(rating_df,books_df,on='ISBN')
df
```

```python
combine_books_rating = df.dropna(axis=0, subset=['Book-Title'])
books_ratingCount = (combine_books_rating.groupby(by=['Book-Title'])['Book-Rating'].count().reset_index().rename(columns={'Book-Rating':'totalRatingCount'})[['Book-Title','totalRatingCount']])
books_ratingCount
```

```python
rating_with_totalRatingCount = combine_books_rating.merge(books_ratingCount, left_on='Book-Title',right_on='Book-Title',how='left')
rating_with_totalRatingCount.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(books_ratingCount['totalRatingCount'].describe())
```

```python
popularity_threshold = 50
rating_popular_books = rating_with_totalRatingCount.query('totalRatingCount >=@popularity_threshold')
rating_popular_books.head()
```

```python
rating_popular_books.shape
```

```python
# private matrix

books_features_df=rating_popular_books.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating').fillna(0)
books_features_df.head()
```

```python
from scipy.sparse import csr_matrix

books_features_df_matrix = csr_matrix(books_features_df.values)

from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', algorithm='brute')
model_knn.fit(books_features_df_matrix)
```

```python
books_features_df.head()
```

```python
query_index = np.random.choice(books_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(books_features_df.iloc[query_index,:].values.reshape(1,-1),n_neighbors=10)
```

```python
books_features_df.head()
```

```python
for i in range(0,len(distances.flatten())):
    if i == 0 :
        print('Recommendation for {0}:\n'.format(books_features_df.index[query_index]))
    else :
        print('{0}:{1}, with distance of {2}:'.format(i,books_features_df.index[indices.flatten()[i]],distances.flatten()[i]))
```
